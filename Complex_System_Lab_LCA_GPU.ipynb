```
D = sio.loadmat('dict512.mat')['Wp'].T
```

```
y = P[0:1024,:]
```

```
N = 300
a = 0.01
adapt = 0.99
lamb = 0.05
```

```
D = torch.tensor(D).float().cuda()
y = torch.tensor(y).float().cuda()
```

```
y.shape,D.shape
```

```
zero = torch.zeros(1,).cuda()
```

```
def LCA3(D,y):

    u = torch.zeros((y.shape[0], D.shape[1])).cuda()

    s = torch.zeros(u.shape).cuda()

    c = torch.matmul(D.T,D) - torch.eye(D.shape[1]).cuda() 

    b = torch.matmul(y,D)

    for i in range(N):

        u = a*(b-torch.matmul(s,c))+(1-a)*u
        
        # s = torch.sign(u)*torch.relu(u)
        s = torch.sign(u)*torch.maximum(zero,torch.absolute(u))
        
    return s.T
```

```
x = LCA3(D,y)
```
